```python
import pandas as pd

# --- Given: load the data (don't change this) ---
url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df = pd.read_csv(url)

print('shape:', df.shape)
df.head()
```

## Task 1 — Missing-value report
For each column, show how many values are missing (and, nice-to-have, the % missing).
<br>Hint: `.isnull()` produces a True/False frame; what does `.sum()` do to booleans? Sort the result so the worst columns are on top.

```python
# TODO: count missing values per column

def missing_summary(df):
    count = df.isna().sum()
    percent = df.isna().mean() * 100
    summary = pd.DataFrame({
        'missing_count': count,
        'missing_percent': percent
    })
    # keep only columns that actually have missing values, worst first
    summary = summary[summary['missing_count'] > 0].sort_values(
        'missing_count', ascending=False
    )
    return summary

print(missing_summary(df))
```

## Task 2 — Survival rates
(a) Overall survival rate (what fraction of passengers survived?).
<br>(b) Survival rate grouped by **passenger class** (`Pclass`).
<br>(c) Survival rate grouped by **sex** (`Sex`).
<br>Hint: `Survived` is 0/1, so the *mean* of that column IS the survival rate. Combine with `.groupby(...)`.

```python
# TODO (a): overall survival rate

df['Survived'].mean()

# TODO (b): survival rate by Pclass
df.groupby('Pclass')['Survived'].mean()

# TODO (c): survival rate by Sex
df.groupby('Sex')['Survived'].mean()

print('Overall survival rate:', df['Survived'].mean())
print('Survival rate by Pclass:', df.groupby('Pclass')['Survived'].mean())
print('Survival rate by Sex:', df.groupby('Sex')['Survived'].mean())
```

## Task 3 — Age: survivors vs non-survivors
Compare the age of passengers who survived vs. those who didn't (e.g. average or median age of each group).
<br>Hint: group by `Survived` and look at `Age`. What does the difference (or lack of one) suggest?

```python
# TODO: age of survivors vs non-survivors

# Average and median age by survival status
df.groupby('Survived')['Age'].mean()
df.groupby('Survived')['Age'].median()

# Or both at once
df.groupby('Survived')['Age'].agg(['mean', 'median'])
```

## Task 4 — Fill missing ages
`Age` has missing values. Fill them in, then confirm there are no missing ages left.
<br>In the cell below, **explain in words**: why fill with the *median* age rather than the *mean*? (Think about outliers / skew.)

```python
# TODO: fill missing Age values with the median, then verify none remain

median_age = df['Age'].median()
df['Age'] = df['Age'].fillna(median_age)

# Confirm no missing values remain
df['Age'].isna().sum()  # Should be 0
```

**Why median, not mean?**

> _your answer here_
The Age distribution is right-skewed — a long tail of older passengers pulls the mean upward, away from where most people actually are. The median ignores those extreme values and reports the middle person, so it's a more representative "typical" age. Filling with it distorts the distribution less.
